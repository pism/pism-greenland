```python
import datetime
from uafgi import dtutil
import scipy.stats

# Identify glaciers with simple retreat
# Plot all the glacier retreats
ax = None
dfg = rh.groupby(['w21t_glacier_number'])

#fig, ax = plt.subplots(1, 1, figsize = (15, 10))

select_df = select.df.set_index('w21t_glacier_number')
data = list()
for count,(w21t_glacier_number,df) in enumerate(dfg):

    # Lookup info in main select.df
    selrow = select_df.loc[w21t_glacier_number]


    # Keep only since year 2000
    df = df[df['date'] > datetime.datetime(2000,1,1)].copy()
#    print(selrow)
#    print(df)

#    df['up_area'] /= df['up_area'].iloc[0]
    df['up_length'] = df['up_area'] / (selrow['w21_mean_fjord_width'] * 1000)
    df['advret'] = df['up_length'] - df['up_length'].iloc[0]
    y = np.array(df['advret'].values, dtype=float)
    x = np.array(df['date'].map(dtutil.year_fraction).values, dtype=float)

#    slope, intercept, rvalue, pvalue, std_err
    res= scipy.stats.linregress(x,y)
#    print('{} {}: {:0.2f} {:0.2f} {:0.2e}'.format(selrow.w21t_Glacier, selrow.ns481_grid, res.slope, res.rvalue, res.pvalue))
    data.append((w21t_glacier_number,res.slope,res.rvalue,res.pvalue))

regdf = pd.DataFrame(data, columns=('w21t_glacier_number', 'slope', 'rvalue', 'pvalue'))
regdf = pdutil.merge_addcols(regdf, select.df, on='w21t_glacier_number', addcols=['w21t_Glacier','ns481_grid'])
regdf.sort_values('slope')
```
